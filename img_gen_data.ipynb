```python
import os
import pandas as pd

# --- Конфигурация ---
IMG_DIR = "img"
GROUND_TRUTH_FILE = "ground_truth.csv"

# --- Логика ---
try:
    image_files = sorted([f for f in os.listdir(IMG_DIR) if f.endswith(('.jpg', '.jpeg', '.png'))])
    print(f"Найдено {len(image_files)} изображений.")
    
    ground_truth_data = []
    for filename in image_files:
        
        # Приводим имя файла к нижнему регистру для надежности
        fn_lower = filename.lower()
        breed = "" # Инициализируем переменную
        
        # Определяем породу по префиксу в имени файла
        if fn_lower.startswith('brit'):
            breed = "Британская короткошерстная"
        elif fn_lower.startswith('rus'):
            breed = "Русская голубая"
        elif fn_lower.startswith('scot'):
            breed = "Шотландская вислоухая"
        else:
            # Все остальные файлы, включая 'random', помечаем как 'Другая'
            breed = "Другая"
            
        ground_truth_data.append({
            "filename": filename,
            "breed": breed
        })
        
    df_truth = pd.DataFrame(ground_truth_data)
    df_truth.to_csv(GROUND_TRUTH_FILE, index=False, encoding='utf-8')
    
    print(f"✅ Файл с правильными ответами '{GROUND_TRUTH_FILE}' успешно создан.")
    print("Пример содержимого:")
    display(df_truth)

except FileNotFoundError:
    print(f"❌ Ошибка: Директория '{IMG_DIR}' не найдена. Не могу создать файл с ответами.")
```
